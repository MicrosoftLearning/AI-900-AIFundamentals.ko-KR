# 대화형 AI

얼마나 자주 인스턴트 메시징, 소셜 미디어, 이메일 또는 기타 온라인 기술을 통해 다른 사람과 커뮤니케이션하는지 생각해 보세요. 많은 이들에게 이것은 바둑판 형태의 접촉입니다. 회사에서 궁금한 점이 있을 때 모바일 디바이스에서 사용할 수 있는 채팅 메시지를 통해 동료와 연락할 수 있으므로 언제나 연결되어 있는 셈입니다.

![대화를 나누고 있는 사람과 로봇](./images/conversational_ai.jpg)

봇은 이러한 종류의 채널을 사용하여 커뮤니케이션하는 AI 상담원으로, 소프트웨어 서비스와 자연스럽게 대화할 수 있게 해줍니다.

## QnA Maker 기술 자료 만들기

고객 지원 시나리오의 경우 웹 사이트 채팅 창, 이메일 또는 음성 인터페이스를 통해 자주 묻는 질문을 해석하고 대답할 수 있는 봇을 만드는 것이 일반적입니다. 봇 인터페이스의 바탕에는 적절한 응답을 위해 봇이 검색할 수 있는 질문과 적절한 답변의 기술 자료가 있습니다.

QnA Maker 서비스는 질의응답 쌍을 입력하거나 기존 문서나 웹 페이지에서 검색하여 빠르게 기술 자료를 만들 수 있도록 해주는 Azure의 Cognitive Service입니다. 그런 다음에 기본 제공 자연어 처리 기능을 사용하여 질문을 해석하고 적절한 답변을 찾을 수 있습니다.

1. 다른 브라우저 탭을 열고 QnA Maker 포털(https://qnamaker.ai)로 이동합니다. Azure 구독과 연결된 Microsoft 계정으로 로그인합니다.
2. QnA Maker 포털에서 **기술 자료 만들기**를 선택합니다.
3. 아직 QnA Maker 서비스 리소스를 만들지 않은 경우 **QnA 서비스 만들기**를 선택합니다. 구독에서 QnA Maker 서비스를 만들 수 있도록 다른 탭에서 Azure Portal이 열립니다. 다음 설정을 사용합니다.
   
   - **구독**: *사용자의 Azure 구독*
   - **리소스 그룹**: *기존 리소스 그룹을 선택하거나 새로 만들기*
   - **이름**: *QnA 서비스의 고유한 이름*
   - **가격 책정 계층**: F0
   **Azure Search 가격 책정 계층**: F
   - **Azure Search 위치**: *사용 가능한 위치*
   - **앱 이름**: _**이름**과 동일(".azurewebsites.net"이 자동으로 추가됨)_
   - **웹 사이트 위치**: _**Azure Search 위치**와 동일_
   - **App Insights**: 사용 안 함

> **참고**: 이미 무료 계층 **QnA Maker** 또는 **Azure Search** 리소스를 프로비전한 경우에는 할당량 때문에 다른 리소스를 만들지 못할 수도 있습니다. 그럴 경우에는 **F0** / **F**가 아닌 계층을 선택하세요.

4. Azure Portal에서 QnA 서비스 및 관련 리소스의 배포가 완료될 때까지 기다립니다.
5. QnA Maker 포털 랩으로 돌아가고, **2단계** 섹션에서 **새로 고침**을 클릭하여 사용 가능한 QnA 서비스 리소스의 목록을 새로 고칩니다.
6. 다음 옵션을 선택하여 QnA 서비스를 KB에 연결합니다.
  - **Microsoft Azure Directory ID**: *구독을 위한 Azure Directory ID*
  - **Azure 구독 이름**: *사용자의 Azure 구독*
  - **Azure QnA 서비스**: *이전 단계에서 만든 QnA 서비스 리소스*
  - **언어**: 영어

### (!) 체크 인
해당 역할에 작업할 수행할 수 있는 권한이 없다는 내용의 메시지가 표시되면 QnA Maker 포털의 브라우저 페이지를 새로 고칩니다.

7. **3단계** 섹션에서 **Margie's Travel KB**라는 이름을 입력합니다.
8. **4단계** 섹션의 **URL** 상자에 *https://github.com/MicrosoftLearning/mslearn-ai900/raw/main/data/qna_bot/margies_faq.docx*를 입력하고 **+ URL 추가**를 클릭합니다. 그런 다음에 **대화** 아래에서 **전문가**를 선택합니다.
9. **5단계** 섹션에서 **KB 만들기**를 클릭합니다.
10. 기술 자료가 생성될 동안 1분 정도 기다립니다. 그런 다음 FAQ 문서에서 가져온 질문과 답변 그리고 전문가 대화의 미리 정의된 응답을 검토합니다.


## 기술 자료 편집

기술 자료는 FAQ 문서의 세부 정보와 몇 가지 미리 정의된 응답을 기반으로 합니다. 이를 보완하기 위해 맞춤형 질의응답 쌍을 추가할 수 있습니다.

1. **+ QnA 쌍 추가**를 클릭합니다.
2. **질문** 상자에 `Hello`를 입력합니다. 그런 다음에 **+ 대체 문구 추가**를 클릭하고 `Hi`을 입력합니다.
3. **답변** 상자에 `Hello`를 입력합니다.

## 기술 자료 학습 및 테스트

이제 기술 자료가 있으므로 QnA Maker 포털에서 테스트할 수 있습니다.

1. 페이지의 오른쪽 상단에서 **저장 및 학습**을 선택하여 기술 자료를 학습시킵니다. 단추를 보려면 창의 크기를 조정해야 할 수도 있습니다.
2. 학습이 완료된 후에 **&larr; 테스트**를 클릭하여 테스트 창을 엽니다.
3. 테스트 창의 하단에서 *안녕* 메시지를 입력합니다. **안녕하세요** 응답이 반환되어야 합니다.
4. 테스트 창의 하단에서 *항공편을 예약하고 싶어요* 메시지를 입력합니다. FAQ에서 적절한 응답이 반환되어야 합니다.
5. 기술 자료의 테스트가 완료되었으면 **&rarr; 테스트**를 클릭하여 테스트 창을 닫습니다.


## 기술 자료를 위한 봇 만들기

기술 자료는 클라이언트 애플리케이션이 일종의 사용자 인터페이스를 통해 질문에 대답하는 데 사용할 수 있는 백엔드 서비스를 제공합니다. 일반적으로, 이러한 클라이언트 애플리케이션은 봇입니다. 봇이 기술 자료를 사용할 수 있도록 하려면 HTTP를 통해 액세스할 수 있는 서비스로 게시해야 합니다. 그런 다음 Azure Bot Service를 사용하여 기술 자료를 바탕으로 사용자 질문에 답변하는 봇을 만들고 호스팅할 수 있습니다.

1. QnA Make 페이지 상단에서 **게시**를 클릭합니다. 그런 다음 **Margies Travel KB** 페이지에서 **게시**를 클릭합니다.
2. 서비스가 배포된 후에 **봇 만들기**를 클릭합니다. 그러면 Azure 구독에서 Web App Bot을 만들 수 있도록 새 브라우저 탭에서 Azure Portal이 열립니다.
3. Azure Portal에서 다음 설정을 사용하여 Web App Bot을 만듭니다(대부분의 설정은 이미 입력되어 있음).
  - **봇 핸들**: *봇의 고유한 이름*
  - **구독**: *사용자의 Azure 구독*
  - **리소스 그룹**: *QnA Maker 리소스를 포함하는 리소스 그룹*
  - **위치**: *QnA Maker 서비스와 동일한 위치입니다*.
  - **가격 책정 계층**: F0
  - **앱 이름**: ***봇 핸들**과 동일하며 *.azurewebsites.net*이 자동으로 추가됨
  - **SDK 언어**: *C# 또는 Node.js 중에서 선택*
  - **QnA 인증 키**: *QnA 기술 자료의 인증 키로 자동 설정되어야 함*
  - **App Service 요금제/위치**: *적합한 요금제 및 위치로 자동 설정되어야 함*
  - **Application Insights**: 꺼짐
  - **Microsoft 앱 ID 및 암호**: 앱 ID 및 암호를 자동으로 만듭니다.
4. 봇이 생성될 때까지 기다립니다. (기다리는 동안 종처럼 생긴 오른쪽 상단의 알림 아이콘이 움직입니다.) 그런 다음에 배포가 완료되었다는 알림에서 **리소스로 이동**을 클릭합니다. (대안으로, 홈 페이지에서 **리소스 그룹**을 클릭하고, 웹앱 봇을 만든 리소스 그룹을 열고, 웹앱 봇을 클릭합니다.)
5. 봇의 왼쪽 창에서 **설정**을 찾고, **웹 채팅에서 테스트**를 클릭하고, 봇에 **안녕하세요. 환영합니다!** 메시지가 표시될 때까지 기다립니다. (봇이 초기화되려면 몇 초 정도 걸릴 수 있습니다.)
6. 테스트 채팅 인터페이스를 사용하여 봇이 기술 자료의 질문에 예상대로 답변하는지 확인합니다. 예를 들어 *호텔을 취소해야 해요*를 제출해 봅니다.


## 채널을 통해 봇에 액세스

봇은 하나 이상의 *채널*을 통해 사용자에게 인터페이스를 제공하는 데 사용될 수 있습니다. 예를 들어 동일한 봇이 웹 채팅 인터페이스, 이메일 및 Microsoft Teams를 통한 상호 작용을 지원할 수 있습니다.

1. 봇의 왼쪽 패널에서 **설정**을 찾고 **채널**을 클릭합니다.
2. **웹 채팅** 채널이 자동으로 추가되었으며 흔히 사용되는 통신 플랫폼용 기타 채널도 사용 가능합니다.
3. **웹 채팅** 채널에서 **편집**을 클릭합니다. 그러면 웹 페이지에 봇을 포함하기 위해 필요한 설정이 있는 페이지가 열립니다. 봇을 포함하려면 제공된 HTML 임베드 코드 그리고 봇에 대해 생성된 비밀 키 하나가 필요합니다.
4. **임베드 코드**를 복사하고 아래 셀에 붙여 넣어 `<!-- EMBED CODE GOES HERE -->` 코멘트를 대체합니다.
5. 비밀 키 하나(어떤 것인지는 중요하지 않음)에 대해 **표시**를 클릭하고, 비밀 키를 복사합니다. 그런 다음에 아래 HTML 임베드 코드에 붙여 넣어 `YOUR_SECRET_HERE`를 대체합니다.
6. HTML 코드의 **최소 높이** 값을 **200px**로 변경합니다(기본값은 500px). 그러면 스크롤하지 않고도 HTML 인터페이스가 보이도록 할 수 있습니다.
7. 셀 왼쪽에 있는 **셀 실행**(&#9655;) 단추를 클릭하여 아래의 셀을 실행함으로써 HTML을 렌더링합니다.
